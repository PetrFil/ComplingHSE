1. Предположим, мы хотим внести ряд изменений на границах слов. Функционал regex позволяет рассматривать границу как 
перед словами, так и в конце слов(\m и \M), в то время как re работает просто с границами слов, независимо 
от их расположения (\b). Например, мы собираемся обозначить границы знаком "_". regex позволяет сделать это так, 
чтобы знака не возникало в начале предложений и не получалось двойных "_" между словами.

```python
import regex, re
text = '''
Не выходи из комнаты, не совершай ошибку.
Зачем тебе Солнце, если ты куришь Шипку?
За дверью бессмысленно все, особенно - возглас счастья.
Только в уборную - и сразу же возвращайся.
'''
a = re.sub(r'\b', '_', text)
b = regex.sub(r'\M', '_', text)
print(a)
print(b)
```

Как мы видим, regex в данном случае позволяет сформулировать более точную задачу.
